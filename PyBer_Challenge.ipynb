# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = 'Resources/city_data.csv'
ride_data_to_load = 'Resources/ride_data.csv'

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how='left', on=['city', 'city'])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_type_count = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_by_type_count
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_type_count = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_by_type_count
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_type_count = pyber_data_df.groupby(['type']).sum()['fare']
total_fares_by_type_count
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares_by_city_type_count = pyber_data_df.groupby(['type']).sum()['fare'] / total_rides_by_type_count
average_fares_by_city_type_count
```

```python
#  4. Get the average fare per ride for each city type (alternate method).
average_fares_by_city_type_count = pyber_data_df.groupby(['type']).mean()['fare']
average_fares_by_city_type_count
```

```python
# 5. Get the average fare per driver for each city type. 
average_fares_by_driver_type_count = pyber_data_df.groupby(['type']).sum()['fare'] / total_drivers_by_type_count
average_fares_by_driver_type_count
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
             'Total Rides': total_rides_by_type_count,
             'Total Drivers': total_drivers_by_type_count,
             'Total Fares': total_fares_by_type_count,
             'Average Fare Per Ride': average_fares_by_city_type_count,
           'Average Fare Per Driver': average_fares_by_driver_type_count})

# Display the DataFrame
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,.0f}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,.0f}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare Per Ride'] = pyber_summary_df['Average Fare Per Ride'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare Per Driver'] = pyber_summary_df['Average Fare Per Driver'].map('${:,.2f}'.format)

# Print the Pyber summary DataFrame
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_by_date_df = pyber_data_df.groupby(['type','date']).sum()['fare']
fare_by_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_data_df.reset_index()
pyber_data_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_date_df = pd.pivot(pyber_data_df, index = 'date', columns = 'type', values = 'fare')
pyber_date_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range_df = pyber_date_df.loc['2019-01-01':'2019-04-29']
date_range_df
```

```python
# 6. Set the 'date' index to datetime datatype. This is necessary to use the resample() method in Step 8.
date_range_df.index = pd.to_datetime(date_range_df.index)
date_range_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_range_df.info()
```

```python
# 8. Create a new DataFrame using the 'resample()' function by week 'W' and get the sum of the fares for each week.
date_week_df = date_range_df.resample('W').sum()
date_week_df
```

```python
date_week_df['Urban']
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Import numpy
import numpy as np

# Define the x axis as the date index and the and y axes as the corresponding data for each of the city types
x_axis = date_week_df.index
y_axis1 = date_week_df['Rural']
y_axis2 = date_week_df['Suburban']
y_axis3 = date_week_df['Urban']

# Plot the data from the DataFrame
fig, ax = plt.subplots(figsize = (20, 5))
ax.plot(x_axis, y_axis1)
ax.plot(x_axis, y_axis2)
ax.plot(x_axis, y_axis3)

# Add a title and x&y labels
ax.set_title('Total Fare by City Type')
#ax.set_xlabel('Date')
ax.set_ylabel('Fare ($USD)')

# Set the horizontal/vertical ticks
ax.set_xticks(['2019-01','2019-02','2019-03','2019-04'])
ax.set_xticklabels(['Jan \n 2019','Feb','March','Apr'])
ax.set_yticks(np.arange(0, 3000, step=500))

# Add a grid
ax.grid(b = True)

# Add a legend
lgd = ax.legend(['Rural', 'Suburban', 'Urban'],
                 fontsize = '15',
                 mode = 'expanded',
                 scatterpoints = 1,
                 loc = 'best',
                 title = 'Type')
lgd.get_title().set_fontsize(15)

# Save the figure
plt.savefig('analysis/Dev2_Plot.png', bbox_inches='tight')

# Display plot
plt.show()
```
